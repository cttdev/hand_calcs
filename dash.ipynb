```python
a = 0.025894434622
l = 0.050800002032

b = 0.0381
h = 0.0015875
I = (1.0 / 12.0) * b * h**3.0

F = 100

E = 69 * 10**9

y = abs(((F * a**2) / (6.0 * E * I)) * (a - 3.0 * l))
print("Deflection: {}".format(y * (1000 / 1)))

M_max = F * a
sigma_max = (M_max * h) / (2.0 * I)

print("Max stress: {} GPa".format(sigma_max * (10**-9)))

F_tuy_L_B = 262000777.14

assert F_tuy_L_B > sigma_max

print("FOS: {}".format(F_tuy_L_B / sigma_max))
```
